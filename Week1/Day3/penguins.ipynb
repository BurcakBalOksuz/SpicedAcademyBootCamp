**MATPLOTLIB**

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

**Build the next plot up gradually**

```python
plt.figure(figsize=(10,5))
plt.scatter(penguins["Culmen Depth (mm)"],penguins["Flipper Length (mm)"],s=100,color="green",marker="o")
plt.xlabel("Culmen Depth in mm",fontsize=15)
plt.xticks()
plt.ylabel("Culmen Length in mm",fontsize=15)
plt.title("Culmen Length and Depth",fontsize=20)
plt.show()
```

**Multiple plots in one**

```python
plt.figure(figsize=(10,5))
plt.style.use("ggplot")
plt.plot(penguins["Culmen Length (mm)"],label="Culmen Length in mm")
plt.plot(penguins["Culmen Depth (mm)"],label="Culmen Depth in mm")
plt.legend()
```
