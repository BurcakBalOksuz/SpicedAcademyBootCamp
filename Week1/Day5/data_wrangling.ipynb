#### 2. Combine the sex data with the rest

```python
#inner merge /join:keeps only the records that exist in both 
pd.merge(penguins_simple_df,penguins_sex_df,how="inner")
```

```python
#outer merge /join:keeps all records that exist in both 
pd.merge(penguins_simple_df,penguins_sex_df,how="outer")
```

```python
#left merge /join:keeps allrecords from the left dataframe 
df=pd.merge(penguins_simple_df,penguins_sex_df,how="left")
df
```

```python
#right merge /join:keeps allrecords from the right dataframe 
pd.merge(penguins_simple_df,penguins_sex_df,how="right")
```
