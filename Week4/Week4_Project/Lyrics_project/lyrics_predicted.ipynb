##  A dataset consisting of Artist name, Lyrics links, Lyrics columns

```python
def get_final_df(df_columns,artists,base,link_initial,headers):
    
    df = pd.DataFrame(columns=df_columns)
    #links_list, artist_list, lyr_list = [],[],[]
    for artist in artists:
        counter=0
        response=requests.get(base+artist,headers)
        time.sleep(1)
        lyrics_html=response.text
        lyrics_soup=BeautifulSoup(lyrics_html)
        print("...counter starts...:")
        for item in lyrics_soup.find_all("strong"):
            if len(item.find_all("a")) >0:
                link_temp = link_initial+item.a.attrs["href"]
                print("counter:",counter,'__','Links:',link_temp)
                #links_list.append(link_temp)
                #artist_list.append(artist)
                df.loc[counter,"Link"]=link_temp
                df.loc[counter,"Artist"]=artist
                content_html=requests.get(link_temp,headers).text
                time.sleep(2)
                content=BeautifulSoup(content_html)
                lyrics_final=''
                if content.find(class_ = 'lyric-body') is not None:
                    content_splitted=re.split('[\n|]',content.find(class_ = 'lyric-body').text)
                    result=[line.replace('\r',"") for line in content_splitted]
                    lyrics_final=[i for i in result if i!='']  
                    #lyr_list.append(lyrics_final)
                    df.loc[counter,"Lyrics"]=lyrics_final         
                else:
                    print('Skipped due to NONE!')
                    continue
                
                counter=counter+1
                if counter>500:
                    break
                file_object = open('sample.csv', 'a')
                for v in lyrics_final:
                    
                    file_object.write(v+";")
                file_object.close()
    
    df.to_csv("final_cold_lyrics_df.csv")
    return df
```

```python
df_columns=["Artist","Link","Lyrics"]
artists=["Coldplay","Radiohead"]
base="https://www.lyrics.com/artist/"
link_initial="https://www.lyrics.com"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
```

```python
df=get_final_df(df_columns,artists,base,link_initial,headers)
df
```
